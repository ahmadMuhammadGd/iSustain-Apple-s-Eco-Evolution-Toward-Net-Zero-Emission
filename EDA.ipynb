**Greenhouse Gas Emissions:**
- Greenhouse gas emissions refer to the release of gases into the Earth's atmosphere that have the potential to trap heat and contribute to global warming and climate change.
- These emissions are typically measured in metric tons of carbon dioxide equivalent (CO2e), which is a standard unit that combines the effect of different greenhouse gases.
- Apple is tracking and reporting their greenhouse gas emissions as part of their commitment to environmental sustainability.

**Fiscal Year:**
- A fiscal year is a period of 12 consecutive months used for financial reporting and accounting purposes.
- Apple's fiscal calendar starts on the last Sunday of September and is 364 days long. This is their chosen timeframe for reporting financial and environmental data.

**Category:**
- Emissions are divided into two categories:
  1. **Corporate Emissions:** These are emissions associated with Apple's corporate operations, such as emissions from their offices, facilities, and corporate activities.
  2. **Product Life Cycle Emissions:** These are emissions associated with the entire life cycle of Apple's products, including manufacturing, transportation, use by customers, and disposal.

**Type:**
- There are two types of emissions data included:
  1. **Gross Emissions:** These are emissions that add to the carbon footprint. In other words, they represent the total emissions produced.
  2. **Carbon Removals:** These are emissions that subtract from the carbon footprint. They typically represent activities that offset or remove an equivalent amount of emissions from the atmosphere, such as carbon offset projects or tree planting.

**Scope:**
- There are three scopes related to greenhouse gas emissions:
  1. **Direct Scope 1 Emissions:** These are emissions from sources owned or controlled by Apple. For example, emissions from Apple-owned manufacturing facilities.
  2. **Indirect Scope 2 Emissions:** These are emissions from purchased electricity, steam, heating, and cooling. This includes emissions indirectly associated with Apple's energy consumption.
  3. **Indirect Scope 3 Emissions:** These are emissions from purchased goods and services, transportation and distribution, business travel, employee commute, product use, and end-of-life activities. It covers emissions related to Apple's entire value chain and product life cycle.

**Description:**
- The description typically provides information about the source or origin of the greenhouse gas emissions. It can help identify what specific activities or processes are contributing to emissions.

**Emissions:**
- This represents the quantity of greenhouse gas emissions, usually measured in metric tons of CO2e, for a particular category, scope, and type of emission. It quantifies the environmental impact of these emissions.

**Carbon Footprint by Product:**
- This section provides data specific to Apple's products and their carbon footprints.

**Release Year:**
- This is the year when a specific Apple product was released to the market.

**Product:**
- The name of the Apple product for which the carbon footprint data is reported.

**Baseline Storage:**
- This refers to the storage capacity of the product with the lowest storage option. It's a characteristic of the product being analyzed.

**Carbon Footprint:**
- This represents the greenhouse gas emissions associated with the product's life cycle, measured in kilograms of CO2e. It quantifies the environmental impact of producing and using the product.

**Normalizing Factors:**
- These are additional data points that can be used to provide context for understanding Apple's environmental and financial performance.

**Revenue:**
- This represents Apple's net sales, typically in millions of US dollars. It's a measure of the company's financial performance.

**Market Capitalization:**
- Market capitalization represents the total value of a company's outstanding shares of stock, typically in billions of US dollars. It provides insight into the company's overall value in the stock market.

**Employees:**
- This indicates the number of full-time equivalent employees working at Apple during the fiscal year. It provides information about the company's workforce size.

In summary, the dataset you have contains detailed information about Apple's greenhouse gas emissions, carbon footprint of products, and additional financial and organizational data. It's used to track Apple's progress towards their goal of becoming carbon neutral by 2030 and provides insights into their environmental and financial performance.

```python
import os
from tabulate import tabulate

# Specify the folder containing CSV files
folder_path = 'apple_emissions'

# Initialize an empty list to store DataFrames
dataframes_list = []

# Iterate over files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Append the DataFrame to the list
        dataframes_list.append(df)

# Access DataFrames by index
for idx, df in enumerate(dataframes_list):
    print(f"\n\n{os.listdir(folder_path)[idx]}:")
    print(tabulate(df, headers='keys', tablefmt='fancy_grid'))
```

```python
scope_dic = {}

scope_dic['Scope 1'] = dataframes_list[1][dataframes_list[1]['Scope'] == 'Scope 1']['Description'].unique()
scope_dic['Scope 2 (market-based)'] = dataframes_list[1][dataframes_list[1]['Scope'] == 'Scope 2 (market-based)']['Description'].unique()
scope_dic['Scope 3'] = dataframes_list[1][dataframes_list[1]['Scope'] == 'Scope 3']['Description'].unique()

for scope, descriptions in scope_dic.items():
    print(f"{scope}:")
    for description in descriptions:
        print(f"  - {description}")
    print("\n")
```

# Null Detection

```python
df.sort_values('Fiscal Year', inplace=True)
# Filter rows with null emissions values
rows_with_null_emissions = df[df['Emissions'].isnull()]
rows_with_null_emissions
```

# Understanding Null Context

## Emissions Nulls

```python
for _, descriptions in scope_dic.items():
    for description in descriptions:
        filtered = df[(df['Description'] == description)]
```

```python
n_rows = len(scope_dic)
fig, axes = plt.subplots(n_rows, 1, figsize=(10, 6 * n_rows))

# Iterate over scopes and descriptions
for i, (scope, descriptions) in enumerate(scope_dic.items()):
    ax = axes[i]
    
    # Iterate over descriptions within the current scope
    for description in descriptions:
        filtered = df[df['Description'] == description]
        
        # Create a line plot on the current axis
        sns.lineplot(data=filtered, x='Fiscal Year', y='Emissions', ax=ax, label=description)
    
    # Set the title and legend for the current subplot
    ax.set_title(scope)
    ax.legend()
    
# Adjust spacing between subplots
plt.tight_layout()

# Show the subplots
plt.show()
```

### Results:
- Scope 1 --> Other's nulls should be handeled where Fiscal Year >= 2017
- Scope 2 --> Steam, heating, and cooling's nulls should be handeled where Fiscal Year >= 2018
- Scope 3 --> Third party cloud's nulls should be handeled where Fiscal Year >= 2018

```python
# Handle null values for Scope 1 where Fiscal Year >= 2017
scope_1_condition = (df['Description'] == 'Other (R&D processes & refrigerant leaks)') & (df['Fiscal Year'] >= 2017)
df.loc[scope_1_condition, 'Emissions'].fillna(method='bfill', inplace=True)

# Handle null values for Scope 2 (Steam, heating, and cooling) where Fiscal Year >= 2018
scope_2_condition = (df['Description'] == 'Steam, heating, and cooling') & (df['Fiscal Year'] >= 2018)
df.loc[scope_2_condition, 'Emissions'].fillna(method='bfill', inplace=True)

# Handle null values for Scope 3 (Third party cloud) where Fiscal Year >= 2018
scope_3_condition = (df['Description'] == 'Third-party cloud (market-based)') & (df['Fiscal Year'] >= 2018)
df.loc[scope_3_condition, 'Emissions'].fillna(method='bfill', inplace=True)


print(tabulate(df[scope_1_condition], headers='keys', tablefmt='fancy_grid'))
print(tabulate(df[scope_2_condition], headers='keys', tablefmt='fancy_grid'))
print(tabulate(df[scope_3_condition], headers='keys', tablefmt='fancy_grid'))
```

# Emissions Reductions by Year Includng Carbon Removals

```python
emissions_by_year=df.groupby('Fiscal Year')['Emissions'].sum().reset_index()
max_emissions_2015 = emissions_by_year[emissions_by_year['Fiscal Year']==2015]['Emissions']
plt.figure(figsize=(10, 10))

ax = sns.lineplot(data=emissions_by_year, x = 'Fiscal Year', y='Emissions')

for _, row in emissions_by_year.iterrows():
    year = row["Fiscal Year"]
    emissions = row["Emissions"]
    percentage_change = round(100+((emissions - max_emissions_2015) / max_emissions_2015 * 100), 2)

    ax.text(x=year, y=emissions, 
            s=f'{str(percentage_change[0])}%', ha='center', va='bottom')

# Display the plot

plt.title('Emissions Reduction')
plt.show()
```
